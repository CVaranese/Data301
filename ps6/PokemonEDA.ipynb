## Question 1

Create as many cells as you need below this one to explore propose, explore and answer your first question.

Were more legendary Pokemon made during later generations?

```python

alt.Chart(data).mark_bar().encode(
    alt.X('Generation:O'),
    alt.Y('count(*)')
).transform_data(
    filter='datum.Legendary'
)
```

```python
alt.Chart(data).mark_bar(stacked='normalize').encode(
    alt.X('Generation:O'),
    alt.Y('count(*)'),
    alt.Color('Legendary')
)
```

Though initially it seems like more legendaries were made only in generations 3, 4, and 5, generation 6 just had fewer Pokemon in general, so there were really just fewer legendaries in generations 1 and 2.

## Question 2

Are some Pokemon types better at physical attack/defense than special attack/defense?

```python
alt.Chart(data).mark_text().encode(
    alt.Color('Type1', scale = alt.Scale(range=['#A6B91A', '#705746', '#6F35FC', '#F7D02C', '#D685AD', '#C22E28', 
                                                '#EE8130', '#A98FF3', '#735797', '#7AC74C', '#E2BF65', '#96D9D6',
                                                '#A8A77A', '#A33EA1', '#F95587', '#B6A136', '#B7B7CE', '#6390F0'])),
    alt.Text('Type1'),
    alt.X('average(Attack)'),
    alt.Y('average(SpAtk)')
)
```

```python
alt.Chart(data).mark_text().encode(
    alt.Color('Type1', scale = alt.Scale(range=['#A6B91A', '#705746', '#6F35FC', '#F7D02C', '#D685AD', '#C22E28', 
                                                '#EE8130', '#A98FF3', '#735797', '#7AC74C', '#E2BF65', '#96D9D6',
                                                '#A8A77A', '#A33EA1', '#F95587', '#B6A136', '#B7B7CE', '#6390F0'])),
    alt.Text('Type1'),
    alt.X('average(Defense)'),
    alt.Y('average(SpDef)')
)
```

```python
alt.Chart(data).mark_text().encode(
    alt.Color('Type1', scale = alt.Scale(range=['#A6B91A', '#705746', '#6F35FC', '#F7D02C', '#D685AD', '#C22E28', 
                                                '#EE8130', '#A98FF3', '#735797', '#7AC74C', '#E2BF65', '#96D9D6',
                                                '#A8A77A', '#A33EA1', '#F95587', '#B6A136', '#B7B7CE', '#6390F0'])),
    alt.Text('Type1'),
    alt.X('average(DefRatio)'),
    alt.Y('average(AtkRatio)')
).transform_data(
    calculate=[alt.Formula('DefRatio', 'datum.Defense/datum.SpDef'),
               alt.Formula('AtkRatio', 'datum.Attack/datum.SpAtk')]
)
```

It looks like, as expected, types like Psychic and Fairy are much better at special stats than physical stats while types like Rock and Fighting are much better at physical stats.
